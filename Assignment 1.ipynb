# Filtering name and city using spacy

```python
import spacy
import en_core_web_sm
model = en_core_web_sm.load()
```

```python
Person = [] 
City = [] 
for x in df_sample['review']:
    line=model(x)
    for entity in line.ents:
        if (entity.label_ == 'GPE'):
            City.append(entity.text)
        elif (entity.label_ == 'PERSON'):
            Person.append(entity.text)
```
